## **1번** - 자격 증명

```python
def solution(data):
    for i in range(0,len(data)):
        data[i] = data[i].replace(' ','').replace('+','1').replace('-','0')
        
    result = ''
    for string in data:
        result += chr(int(string, 2))
        
    return result
```

```python
test1 = [' + - - + - + - ', ' + + + - + - + ', ' + + - + + + - ']
solution(test1)
```

```python
test2 = [' + + + - - + + ', ' + + + - + - - ', '++----+', '+++ --+ -', '+++-+ - -']	
solution(test2)
```

```python
test3 = [' + + - - - - + ', ' + + - + + - - ', '+ +-- +++ ', ' ++- ++++']
solution(test3)
```

# **2번** - 암호문

```python
import re

def solution(data):
    sum = 0
    m = re.findall('([rev])(10|[0-9])', data)
    
    for i, j in m:
        sum += int(j)
    sum = str(sum)
    
    return f'{sum[0]}월 {sum[1]}일'

test_data = 'a10b9r1ce33uab8wc918v2cv11v9'
solution(test_data)
```

```python
%%timeit

def solution(data):
    result = re.findall(r'([rev])(10|[0-9])', data)
    sum = 0
    for _, j in result:
        sum += int(j)
    return str(sum//10) + '월' + str(sum%10) + '일'

solution('a10b9r1ce33uab8wc918v2cv11v9')
```

```python
%%timeit

def solution(data):
    s = 0
    data = re.findall('([rev])(10|[0-9])', data)
    for _, i in data:
        s += int(i)
    return f'{str(s)[0]}월 {str(s)[1]}일'

solution('a10b9r1ce33uab8wc918v2cv11v9')
```

# 6번 - 샌드위치 포장

```python
def solution(data):
    stack = []
    count = 0
    for i in data:
        stack.append(i)
        if stack[-5:] == [1,2,3,4,1]:
            count += 1
            for _ in range(5):
                stack.pop()
    return count
```

```python
test1 = [1, 2, 3, 4, 1, 1, 2, 3, 4]
solution(test1)
```

```python
test2 = [1, 1, 1, 2, 3, 4, 1, 2, 3, 4, 1]
solution(test2)
```

```python
test3 = [1, 2, 3, 4, 2, 3, 4, 1]
solution(test3)
```
