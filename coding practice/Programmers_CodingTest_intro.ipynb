# Ïò∑Í∞ÄÍ≤å Ìï†Ïù∏ Î∞õÍ∏∞

```python
def solution(price):
    discount = (20 if price >= 500000 else 10 if price >= 300000 else 5 if price >= 100000 else 0) / 100
    return int(price * (1-discount))

solution(150000)
```

# ÌäπÏ†ï Î¨∏Ïûê Ï†úÍ±∞ÌïòÍ∏∞

```python
def solution(my_string, letter):
    s = ''
    for l in my_string:
        s += l if l!=letter else ''
    return s

solution("abcdef", "f")
```
